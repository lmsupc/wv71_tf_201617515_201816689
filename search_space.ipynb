#DEFINICIÓN DEL ESPACIO DE BÚSQUEDA, ESTADO INICIAL, POSIBLE ESTADO FINAL Y TRANSICIONES

##ESPACIO DE BÚSQUEDA
Se empleará una matriz de 90 x 90 para representar la ciudad donde se ubicarán los almacenes y puntos de entrega.


imagen.jpg

##ESTADO INICIAL
El vehículo no debe encontrarse fuera del almacén.

Imagen 1.jpg

##POSIBLES ESTADOS FINALES
El vehículo habrá llegado a uno de los puntos de entrega más cercano utilizando el camino más corto. Se está considerando que cada una de las cuadras tiene la misma distancia.

Imagen 2.jpg

##TRANSICIONES
* Moverse a la **DERECHA**
* Moverse a la **IZQUIERDA**
* Moverse a la **ARRIBA**
* Moverse a la **ABAJO**

##REGLAS
* El vehículo no debe salirse de la matriz antes mencionada.
* El vehículo no debe pasar por otros almacenes.
* El vehículo no debe pasar por puntos de entrega donde no le corresponde.
* El vehículo no debe regresar sobre el camino ya recorrido.
